Step-4:- Visualizing and preprocessing the data for creating storyboards

```python
#visualizing the coloums
data1.columns.values
     
```

```python
#creating a table for better visualization
df=data1[['gname','country_txt','nkill']]
df=df.groupby(['gname','country_txt'],axis=0).sum().sort_values('nkill',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
#setting the new data
data1.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)

#Filter only required columns
data=data1[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
data['casualities']=data['Killed']+data['Wounded']
data.head()
```

```python
#dimensions of new data
data.shape
     
```

```python
#information of new data
data.info()
     
```

```python
#describing the data
data.describe()
```

```python
#saving the data locally for using in tablaeu dashboard
data.to_csv('terror.csv')
```

Plot for casualties based of cities and type of attack,target,weapons

```python
#Top 10
Top_cities = data.groupby('city')['casualities'].max().sort_values(ascending = False).reset_index().head(10)
Top_attackType = data.groupby('AttackType')['casualities'].sum().sort_values(ascending = False).reset_index().head(10)
Top_Target_type = data.groupby('Target_type')['casualities'].sum().sort_values(ascending = False).reset_index().head(10)
Top_Weapon_type = data.groupby('Weapon_type')['casualities'].sum().sort_values(ascending = False).reset_index().head(10)

plt.figure(figsize=(18,15))

plt.subplots_adjust(wspace = .5)
plt.subplot(2,2,1)
sns.barplot(x='casualities', y='city', data = Top_cities)

plt.subplot(2,2,2)
sns.barplot(x='casualities', y='AttackType' , data = Top_attackType)

plt.subplot(2,2,3)
sns.barplot(x='casualities', y= 'Target_type', data = Top_Target_type)

plt.subplot(2,2,4)
sns.barplot(x='casualities', y= 'Weapon_type', data = Top_Weapon_type)
```
